```python
%matplotlib inline
```

```python
# import dependencies

import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create the plot
plt.plot(x_axis, y_axis)
```

```python
# create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston', marker="*", color="blue", linewidth="2")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# create the plot
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, color="green", marker="D", linewidth="2", label="Boston")

ax.legend()

#titles/labels
ax.set_title("PyBer Fare By Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

ax.set_ylim(0, 50)
ax.grid()
```

```python
# create the v bar plot
plt.bar(x_axis, y_axis, color="green", label="Boston")

# titles/labels
plt.title("PyBer Fare by Month")
plt.xlabel("Date")
plt.ylabel("Fare($)")

plt.legend()
```

```python
# create the h bar plot
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()

# titles/labels
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")

plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# create a h plot with ax.plot()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color='cyan', label='Chicago')
ax.legend()

# titles/labels
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare($)')
ax.set_ylabel("Date")

ax.invert_yaxis()
```

```python
# create a scatter plot 1
plt.plot(x_axis, y_axis, 'o')
```

```python
# create a scatter_plot 2
plt.scatter(y_axis, x_axis, color="red", label="Chicago")
plt.legend()

# title/labels
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")

plt.xlim(0, 50)

plt.gca().invert_yaxis()
```

```python
# bubble plots
plt.scatter(x_axis, y_axis)
```

```python
# bubble plots
upsized_yaxis = []
for data in y_axis:
    upsized_yaxis.append(data*3)

plt.scatter(x_axis, y_axis, s=upsized_yaxis)
```

```python
# bubble plots
plt.scatter(x_axis, y_axis, s = [dt_pt * 3 for dt_pt in y_axis])
```

```python
# scatter plots - OOP
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# bubble plots - OOP
fig, ax = plt.subplots()
ax.scatter(y_axis, 
           x_axis, 
           s=[dt_pt * 5 for dt_pt in y_axis], 
           color='skyblue', 
           alpha=0.8, 
           edgecolor="black", 
           linewidth=2, 
           label='Boston')

ax.legend()

# title/labels
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")

ax.set_xlim(0, 50)

ax.invert_yaxis()
```

```python
# pie chart
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# pie chart w/ filters
plt.subplots(figsize=(8, 8))
explode_vals = [0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0]
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.pie(
    y_axis, 
    labels=x_axis, 
    explode=explode_vals, 
    colors=colors,
    autopct="%.1f%%")

plt.show()
```

```python
# pieplot - OOP
fig, ax = plt.subplots(figsize=(8, 8))
explodes = [0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0]
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis, 
       labels=x_axis, 
       autopct="%.1f%%", 
       explode=explodes, 
       shadow=True,
       startangle=95,
       counterclock=False,
       colors=colors
      )

plt.show()
```
